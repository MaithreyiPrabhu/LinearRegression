```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm

%matplotlib inline
```

```python
data = pd.read_csv('Advertising.csv', index_col=0)
```

```python
data
```

## Simple Linear regression

```python
## Seeing effect of TV on sales

plt.figure(figsize=(16,8))
plt.scatter(data['TV'], data['sales'], c='black')
plt.xlabel('Money spend on TV advertisements')
plt.ylabel('Sales in 1000 $')
```

```python
## Seeing effect of radio on sales

plt.figure(figsize=(16,8))
plt.scatter(data['radio'], data['sales'], c='black')
plt.xlabel('Money spend on TV advertisements')
plt.ylabel('Sales in 1000 $')
```

```python
## Seeing effect of newspaper on sales

plt.figure(figsize=(16,8))
plt.scatter(data['newspaper'], data['sales'], c='black')
plt.xlabel('Money spend on TV advertisements')
plt.ylabel('Sales in 1000 $')
```

```python
## TV and sales

#Convert to 2D numpy array
X = data['TV'].values.reshape(-1,1)
Y = data['sales'].values.reshape(-1,1)

#Call the regression model
regression = LinearRegression()

#Fit the model to data
regression.fit(X,Y)


print(regression.intercept_[0], regression.coef_[0][0])
## We get constat of 7.03 and a slope of 0.0475
## Positive constant and positive slope

predictions = regression.predict(X)
plt.figure(figsize=(16,8))
plt.scatter(X,Y, c='black')
plt.xlabel('Money spend on TV advertisements')
plt.ylabel('Sales in 1000 $')
plt.plot(X, predictions, c='blue', linewidth=2)
```

```python
## Assess the quality of the model
X = data['TV']
Y = data['sales']
linearmodel = sm.add_constant(X)
#Estimator Orinary least square
estimator = sm.OLS(Y, linearmodel).fit()
print(estimator.summary())

## Report reading
## R squared value of 0.612 is acheived. Only 60% of the variablity is explained. Not very good.
## F-statisctics is 312 which is much larger than 1. Seems the model is a little good.
```

### Multiple Linear Regression

```python
## Take all features. The news paper, radio and TV on sales data
Y = data['sales'].values.reshape(-1,1)
X_features = data.drop(['sales'], axis=1)


#Call the regression model
regression = LinearRegression()

#Fit the model to data
regression.fit(X_features,Y)


print(regression.intercept_[0])
## We get constat of 7.03 and a slope of 0.0475
```

```python
## Assess the quality of the model
X = np.column_stack((data['TV'], data['radio'],data['newspaper']))
Y = data['sales'].values.reshape(-1,1)
linearmodel = sm.add_constant(X)
estimator = sm.OLS(Y,linearmodel).fit()
print(estimator.summary())
## R squared is 0.897 which is good
## F-statistic is 570 which is very much higher than 1
## But here p value for newspaper is 0.860 which is higher than 0.05 which means newspaper is not relevant for our model and you can take it out
```

```python
predictions = regression.predict(X)
plt.figure(figsize=(16,8))
plt.scatter(X_features,Y, c='black')
plt.xlabel('Money spend on newspaper advertisements')
plt.ylabel('Sales in 1000 $')
plt.plot(X, predictions, c='blue', linewidth=2)
```
